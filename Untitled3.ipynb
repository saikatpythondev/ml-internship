```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
file.replace({"Male":1,"Female":0},inplace=True)
```

```python
file.isna().sum()
```

```python
print(file.corr())
sns.heatmap(file.corr())
```

```python
from sklearn.cluster import KMeans
```

```python
km=KMeans(n_clusters=3)
km.fit(file)
```

```python
err=[]
num=[]
for i in range(1,10):
    km=KMeans(n_clusters=i)
    km.fit(file)
    err.append(km.inertia_)
    num.append(i)
```

```python
plt.plot(num,err)
```

```python
km=KMeans(n_clusters=6)
km.fit(file)
```

```python
pred=km.predict(file)
```

```python
file["predict"]=pred
```

```python
file
```

```python
sns.scatterplot(x=file["Age"],y=file["Spending Score (1-100)"],hue=file["predict"])
```

```python
sns.scatterplot(x=file["Annual Income (k$)"],y=file["Spending Score (1-100)"],hue=file["predict"])
```

```python
from sklearn.model_selection import train_test_split
```

```python
dependent=file.drop("predict",axis=1)
independent=file["predict"]
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(dependent,independent,test_size=0.2,random_state=True)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier()
```

```python
dt.fit(xtrain,ytrain)
```

```python
pre=dt.predict(xtest)
```

```python
dt.score(xtest,ytest)
```
